```python
file = open('input_day2.txt','r')
podatki = file.read().splitlines()

split_data = []

for game in podatki:
    # Split by semicolon and then iterate over each part
    parts = game.split(';')
    game_parts = []
    for part in parts:
        # Split each part by comma
        subparts = part.split(',')
        # Trim whitespace and add to the game_parts list
        game_parts.append([subpart.strip() for subpart in subparts])
    split_data.append(game_parts)

split_data
```

```python
def find_max_values(data):
    max_values = []
    for game in data:
        # Split the game data into parts, ignoring the game title
        parts = game.split(': ')[1].split(';')
        game_max = {'red': 0, 'green': 0, 'blue': 0}
        for part in parts:
            # Split each part by comma and process each item
            items = part.split(',')
            for item in items:
                # Extract the number and color
                number, color = item.strip().split()
                number = int(number)
                # Update the max value for each color
                if number > game_max[color]:
                    game_max[color] = number
        max_values.append(game_max)
    return max_values

max_values_per_game = find_max_values(podatki)
max_values_per_game
```

```python
max_possible_values = {'red': 12, 'green': 13, 'blue': 14}

def sum_possible_game_ids(max_values_per_game, max_possible_values):
    possible_game_ids_sum = 0
    for i, game_max in enumerate(max_values_per_game, start=1):
        if all(game_max[color] <= max_possible_values[color] for color in game_max):
            possible_game_ids_sum += i
    return possible_game_ids_sum

sum_of_possible_ids = sum_possible_game_ids(max_values_per_game, max_possible_values)
sum_of_possible_ids

#sum_of_ids
```
