```python
def calculate_area(directions):
    # Initial position
    x, y = 0, 0
    # Store the vertices of the polygon
    vertices = [(x, y)]

    # Mapping directions to coordinate changes
    direction_map = {
        'R': (1, 0),  # right
        'L': (-1, 0), # left
        'U': (0, 1),  # up
        'D': (0, -1)  # down
    }

    # Trace the path
    for direction, steps, _ in directions:
        dx, dy = direction_map[direction]
        for _ in range(steps):
            x += dx
            y += dy
            vertices.append((x, y))

    # Calculate area using the Shoelace formula
    area = 0
    n = len(vertices)
    for i in range(n - 1):
        area += vertices[i][0] * vertices[i + 1][1]
        area -= vertices[i + 1][0] * vertices[i][1]
    area += vertices[-1][0] * vertices[0][1] - vertices[0][0] * vertices[-1][1]

    return abs(area) / 2

calculate_area(podatki3)
```
