Advent of Code Day 4

```python
file = open('input_day4.txt','r')
podatki = file.read().splitlines()

podatki2 = []
for card in podatki:
    card_number, stevilke = card.split(': ')
    dobitne, izbrane = stevilke.split(' | ')
    dobitne_st = [int(st) for st in dobitne.split()]
    izbrane_st = [int(st) for st in izbrane.split()]
    podatki2.append([card_number, dobitne_st, izbrane_st])

points = 0
for card in podatki2:
    #print(card[1])
    dobitki = list(set(card[1]).intersection(card[2]))
    if len(dobitki) > 0:
        points += 2**(len(dobitki)-1)
print(points)
```

```python
kartice = [1]*len(podatki2)
j = 1
for card in podatki2:
    dobitki = list(set(card[1]).intersection(card[2]))
    for i in range(j, j + len(dobitki)):
        kartice[i] += kartice[j-1]
    print(kartice)
    j += 1
sum(kartice)
#kartice
```
