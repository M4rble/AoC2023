Advent of Code Day 9

```python
file = open('input_day9.txt','r')
podatki = file.read().splitlines()
podatki2 = []
for row in podatki:
    st = row.split(' ')
    stevilo = [int(s) for s in st]
    podatki2.append(stevilo)
podatki2
```

```python
def poisci_razlike(sez):
    return([j-i for i, j in zip(sez[:-1], sez[1:])])

vsota = 0
for row in podatki2:
    seznam_razlik = []
    seznam_razlik.append(row)
    razlike = poisci_razlike(row)
    while any(razlike) != 0:
        seznam_razlik.append(razlike)
        razlike = poisci_razlike(seznam_razlik[-1])
    vsota += sum(s[-1] for s in seznam_razlik)
print(vsota)
```

```python
vsota = 0
for row in podatki2:
    seznam_razlik = []
    seznam_razlik.append(row)
    razlike = poisci_razlike(row)
    while any(razlike) != 0:
        seznam_razlik.append(razlike)
        razlike = poisci_razlike(seznam_razlik[-1])
    vsota += sum(s[0] for s in seznam_razlik[::2]) - sum(s[0] for s in seznam_razlik[1::2])
print(vsota)
```
