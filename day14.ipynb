```python
file = open('input_day14.txt','r')
podatki = file.read()
groups = podatki.strip().split('\n')
#processed_groups = [group.split('\n') for group in groups]
podatki2 = [list(row) for row in groups]
podatki2
```

```python
def move_O_shapes_up(data):

    rows, cols = len(data), len(data[0])

    for col in range(cols):
        for row in range(rows):
            if data[row][col] == 'O':
                # Move the 'O' upwards until a blockage ('#') or another 'O' is encountered, or the top is reached
                current_row = row
                while current_row > 0 and data[current_row - 1][col] == '.':
                    # Swap '.' with 'O'
                    data[current_row][col], data[current_row - 1][col] = data[current_row - 1][col], data[current_row][col]
                    current_row -= 1

    # Convert the list of lists back to the original format
    return [''.join(row) for row in data]

novi_podatki = move_O_shapes_up(podatki2)
```

```python
def count_O_in_rows(data):
    return [row.count('O') for row in data]

# Count the number of 'O's in each row of the moved_data
O_counts = count_O_in_rows(novi_podatki)

vsota = 0
for i in range(len(O_counts),0,-1):
    vsota += i*O_counts[100-i]
print(vsota)
```
