```python
# part2

def find_result_2(list):
    vsota = 0
    skatle = [[] for _ in range(256)]
    extracted = [item.split('=')[0] if '=' in item else item.split('-')[0] for item in list]
    number = [item.split('=')[1] if '=' in item else item.split('-')[1] for item in list]
    signs = [item[item.find('=')] if '=' in item else item[item.find('-')] for item in list]
    for i, sign in enumerate(signs):
        #print(extracted[i], signs[i], number[i])
        cv = 0
        char = [s for s in extracted[i]]
        for c in char:
            cv += ord(c)
            cv *= 17
            cv = cv % 256
        if sign == '-':
            # Remove the sublist where the first element matches extracted[i]
            skatle[cv] = [sublist for sublist in skatle[cv] if sublist[0] != extracted[i]]
        elif sign == '=':
            matched = False
            # Check if there is a sublist starting with extracted[i]
            for sublist in skatle[cv]:
                if sublist[0] == extracted[i]:
                    # Replace the second element of the matched sublist
                    sublist[1] = int(number[i])
                    matched = True
                    break
            if not matched:
                # Append new sublist if no match was found
                skatle[cv].append([extracted[i], int(number[i])])
        #print(skatle)
    for j in range(len(skatle)):
        if len(skatle[j]) > 0:
            for k in range(len(skatle[j])):
                vsota += (j+1) * (k+1) * skatle[j][k][1]
    return vsota
        


find_result_2(podatki)     
```
