```python
file = open('input_day13.txt','r')
podatki = file.read()

groups = podatki.strip().split('\n\n')
processed_groups = [group.split('\n') for group in groups]

processed_groups[0]
```

```python
def find_identical_lines_or_columns(groups):
    results = {}

    for i, group in enumerate(groups):
        # Check for identical lines
        line_pairs = [(idx1, idx2) for idx1, line1 in enumerate(group) 
                      for idx2, line2 in enumerate(group) if line1 == line2 and idx1 < idx2]

        # Transpose the group for column-wise comparison
        transposed_group = [''.join(line[idx] for line in group) for idx in range(len(group[0]))]

        # Check for identical columns
        column_pairs = [(idx1, idx2) for idx1, col1 in enumerate(transposed_group) 
                        for idx2, col2 in enumerate(transposed_group) if col1 == col2 and idx1 < idx2]

        # Store results
        if line_pairs or column_pairs:
            results[i] = {
                "identical_lines": line_pairs,
                "identical_columns": column_pairs
            }

    return results

print(find_identical_lines_or_columns(processed_groups))
processed_groups[4]
```

```python
def find_extended_adjacent_pairs(groups):
    results = {}
    vsota = 0

    def check_extended_pairs(group, idx1, idx2):
        """ Check for extended pairs in the group starting from (idx1, idx2). """
        pairs = [(idx1, idx2)]
        n = len(group)

        # Check for extended pairs going up
        u_idx1, u_idx2 = idx1, idx2
        while u_idx1 > 0 and u_idx2 < n - 1 and group[u_idx1 - 1] == group[u_idx2 + 1]:
            u_idx1 -= 1
            u_idx2 += 1
            pairs.append((u_idx1, u_idx2))

        if len(pairs) < idx1 + 1 and len(pairs) < n - idx2:
            pass
        else:
            return pairs

    for i, group in enumerate(groups):
        # Find adjacent identical lines
        adjacent_lines = [(idx, idx + 1) for idx in range(len(group) - 1) if group[idx] == group[idx + 1]]

        # Find extended adjacent pairs for lines
        extended_lines = [check_extended_pairs(group, idx1, idx2) for idx1, idx2 in adjacent_lines if check_extended_pairs(group, idx1, idx2) is not None]

        # Transpose the group for column-wise comparison
        transposed_group = [''.join(line[idx] for line in group) for idx in range(len(group[0]))]

        # Find adjacent identical columns
        adjacent_columns = [(idx, idx + 1) for idx in range(len(transposed_group) - 1) if transposed_group[idx] == transposed_group[idx + 1]]

        # Find extended adjacent pairs for columns
        extended_columns = [check_extended_pairs(transposed_group, idx1, idx2) for idx1, idx2 in adjacent_columns if check_extended_pairs(transposed_group, idx1, idx2) is not None]

        # Store results
        if extended_lines or extended_columns:
            results[i] = {
                "extended_adjacent_identical_lines": extended_lines,
                "extended_adjacent_identical_columns": extended_columns
            }
        if not extended_lines:
            pass
        else:
            vsota += 100*(extended_lines[0][0][0]+1)
        if not extended_columns:
            pass
        else:
            vsota += extended_columns[0][0][0]+1
    print(vsota)
    return results

find_extended_adjacent_pairs(processed_groups)
```
